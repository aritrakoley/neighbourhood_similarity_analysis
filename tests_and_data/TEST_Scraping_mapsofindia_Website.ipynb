```python
import requests
import pandas as pd
```

```python
#User Input:
city_name = "Kolkata"
city_state = "West Bengal"

#Formatting user input
city_name = city_name.lower().replace(' ', '-')
city_state = city_state.lower().replace(' ', '-')
print(city_name, city_state)
```

```python
raw_page = requests.get("https://www.mapsofindia.com/pincode/india/{}/{}".format(city_state, city_name))
page_html =  raw_page.text
```

```python
ts = page_html.find("<table><tr>")
te = page_html.find("</table>", ts) + 8
table = page_html[ts:te]
```

```python
#SANITY CHECK: rendering the read html table
from IPython.display import display, HTML
chart = HTML(table)
display(chart)  
```

```python
df_list = pd.read_html(table, header=[1], skiprows=0)
print("DataFrames read:", len(df_list))
df = df_list[0]
print("Read Dataframe:")
display(df.head())
print("Rows: ", df.shape[0], "\nCols: ", df.shape[1])
df.dtypes
```
